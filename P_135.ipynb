```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('star_with_gravity C-134.csv')
df.head()
```

```python
mass=df['Mass'].tolist()
radius=df['Radius'].tolist()
gravity=df['Gravity'].tolist()
star_name=df['Star_name'].tolist()
distance=df['Distance'].tolist()

plt.figure(figsize=(10,5))
plt.title('Stars Solar Mass')
plt.bar(star_name[0:9],mass[0:9])
```

```python
plt.figure(figsize=(10,5))
plt.title('Stars Solar Radius')
plt.bar(star_name[0:9],radius[0:9])
```

```python
plt.figure(figsize=(10,5))
plt.title('Stars Solar Gravity')
plt.bar(star_name[0:9],gravity[0:9])
```

```python
plt.figure(figsize=(10,5))
plt.title('Stars Solar Distance')
plt.bar(star_name[0:9],distance[0:9])
```
